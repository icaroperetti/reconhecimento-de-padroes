```python
import pandas as pd
from IPython.display import display

# Cancer Bheavior Risk dataset
url = "https://raw.githubusercontent.com/icaroperetti/reconhecimento-de-padroes/main/cervical-cancer.csv?token=GHSAT0AAAAAABW52AURIKEZM3XKSRTANGRGYYQWPXQ"

df = pd.read_csv(url, sep=",")

display(df)
```

```python
columns = df.columns[0:19]

cancer = df["ca_cervix"]
data = df[columns]

display(data)
```

```python
# Calcular média e variância
for c in columns:
  print("Coluna:",c)
  print("Variância:", data[c].var())
  print("Média:",data[c].mean(),'\n')


data = data[[column for column in columns if column !=  "norm_significantPerson" or column != "behavior_sexualRisk" or column != "attitude_consistency" ]]
```

```python
# Desvio Padrão
dp = data.apply(lambda x: (x-x.mean())/ x.std(), axis=0)
display(dp)
```

```python
# Normalizar entre 0 e 1
normalized_data_zero_and_one=(data-data.min())/(data.max()-data.min())
display(normalized_data_zero_and_one)
```

```python
# Pelo máximo
normalized_max = data / data.max()
display(normalized_max)
```
